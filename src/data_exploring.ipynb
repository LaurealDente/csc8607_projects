```python
from datasets import load_dataset
import pandas as pd

tiny_imagenet = load_dataset("zh-plus/tiny-imagenet")
print(tiny_imagenet)
```

```python

train = tiny_imagenet["train"]
val = tiny_imagenet["valid"]

print(train)
print(val)

print(train.features)
print(val.features)
```

```python
print(train.unique("label"))
print(val.unique("label"))
```

```python
from collections import Counter
label_train_counts = Counter(train["label"])
label_val_counts = Counter(val["label"])

print(label_train_counts)
print(label_val_counts)
```

```python
from matplotlib import pyplot

pyplot.hist(label_train_counts.values())
pyplot.xlabel('nombre de labels')
pyplot.ylabel('nombre de classes')
pyplot.title('Histogramme du nombre de rcurrence des labels')
pyplot.show()


pyplot.hist(label_val_counts.values())
pyplot.xlabel('nombre de labels')
pyplot.ylabel('nombre de classes')
pyplot.title('Histogramme du nombre de rcurrence des labels')
pyplot.show()
```

```python
from PIL import Image
import numpy as np

def images_equal(img1, img2):
    return np.array_equal(np.array(img1), np.array(img2))

matches = [images_equal(img1, img2) for img1, img2 in zip(df_train.iloc[:,0], df_val.iloc[:,0])]

true_indices = [i for i, match in enumerate(matches) if match]

print(true_indices)
```
